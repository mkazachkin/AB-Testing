Гипотезы:
1. Упрощенная форма регистрации пользователя увеличит конверсию в регистрации. Метрика (для метода ICE): конверсия в зарегистрированного пользователя

3. Реорганизация каталога увеличит конверсию пользователей, которые воспользовались каталогом. Метрика (для метода ICE): конверсия в пользователя, воспользовавшегося каталогом
4. Перенастройка выдачи результатов поиска повысит конверсию в покупку. Метрика (для метода ICE): конверсия в покупку
5. Размещение информации о возможности кредитования и его условиях в верхней части страницы увеличит конверсию в продажи. Метрика (для метода ICE): конверсия в покупку

**Сделайте приоритезацию гипотез из предыдущего урока с помощью ICE**

Гипотезы:
1. Упрощенная форма регистрации пользователя увеличит конверсию в регистрации. Метрика (для метода ICE): конверсия в зарегистрированного пользователя
2. Выделение формы поиска с помощю анимированного или яркого элемента увеличит конверсию в использование поиска. Метрика (для метода ICE): конверсия пользователя, воспользовавшегося поиском
3. Реорганизация каталога увеличит конверсию пользователей, которые воспользовались каталогом. Метрика (для метода ICE): конверсия в пользователя, воспользовавшегося каталогом
4. Перенастройка выдачи результатов поиска повысит конверсию в покупку. Метрика (для метода ICE): конверсия в покупку
5. Размещение информации о возможности кредитования и его условиях в верхней части страницы увеличит конверсию в продажи. Метрика (для метода ICE): конверсия в покупку

```python
rice_reach = [50, 100, 80, 50, 100]         # Охват. Зададим в процентах от пользователей. Данные оценочные
rice_impact = [2, 3, 3, 3, 2]               # Влияние. Максимальное влияние - 3
rice_confidence = [0.9, 0.5, 0.8, 1, 0.8]   # Уверенность в оценке. Максимальная уверенность - 1
rice_effort = [1, 0.25, 3, 1, 0.25]          # Трудозатраты. В человеко-месяцах Данные оценочные

ice_impact = [10, 7, 8, 9, 5]               # Влияние на ключевой показатель
ice_confidence = [9, 5, 8, 10, 8]           # Уверенность
ice_ease = [3, 10, 1, 3, 9]                 # Легкость

ice = []
rice = []
for hypothesis_n in range(5):
    rice.append(f'RICE score = ' + str(int(rice_reach[hypothesis_n] * rice_impact[hypothesis_n] 
                                       * rice_confidence[hypothesis_n] / rice_effort[hypothesis_n])).zfill(3) + f'. Гипотеза {hypothesis_n + 1}.')
    ice.append(f'ICE score = ' + str(ice_impact[hypothesis_n] * ice_confidence[hypothesis_n] * 
                                     ice_ease[hypothesis_n]).zfill(3) + f'. Гипотеза {hypothesis_n + 1}.')
for hypothesis_s in sorted(ice, reverse=True):
    print (hypothesis_s)
print ()
for hypothesis_s in sorted(rice, reverse=True):
    print (hypothesis_s)
```

По показателю ICE гипотезы распределились следующим образом:
1. ICE = 360. Размещение информации о возможности кредитования и его условиях в верхней части страницы увеличит конверсию в продажи.
2. ICE = 350. Выделение формы поиска с помощю анимированного или яркого элемента увеличит конверсию в использование поиска. Метрика (для метода ICE): конверсия пользователя, воспользовавшегося поиском.
3. ICE = 270. Перенастройка выдачи результатов поиска повысит конверсию в покупку.
3. ICE = 270. Упрощенная форма регистрации пользователя увеличит конверсию в регистрации.  
5. ICE = 64. Реорганизация каталога увеличит конверсию пользователей, которые воспользовались каталогом.

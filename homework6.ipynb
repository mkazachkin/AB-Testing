# Задача 1
Продакт на главной mail.ru решил протестировать в рекомендательной ленте контента вместо карточек со статьями видеоплеер с короткими видео. Нынешний таймспент на юзера в день в среднем равен 25 минут, а стандартная ошибка (SD) равна 156. Мы предполагаем, что в новой версии таймспент на юзера в день изменится на 10%. Средний трафик 20000 человек в день. Посчитайте сколько дней необходимо держать эксперимент при alpha = 5% и beta = 20%.

## Решение

Среднее время, проводимое пользователем на сайте с карточками составляет 25 минут. Рассчитаем среднее время при uplift = 10 %

```python
time_spent_1 = 25
uplift = 0.1
time_spent_2 = round(time_spent_1 * (1 + uplift), 2)
print(time_spent_2)
```

Ожидаемое среднее время составит 27,5 минут. Рассчитаем размер выборки при $ \sigma = 156, \alpha = 0.05, \beta = 0.20$ Для расчета времени проведения A/B теста расчитаем необходимый объем выборок для каждой группы. Для этого воспользуемся онлайн калькулятором [Sample Size](https://statulator.com/SampleSize/ss2M.html "Sample Size Calculator for Comparing Two Independent Means") для сравнения двух средних:

![Расчет размера выборки](pics/pic_06_01.png "Расчет размера выборки")

Расчет показал, что объем выборки для каждой из групп должен составлять 61124 пользователя, или 122248 пользователей всего. Рассчитаем продолжительность A/B-теста с учетом средней посещаемости в 20000 пользователей в сутки:

```python
import numpy as np

print(np.ceil(122248 / 20000))
```

## Ответ
При заданных параметрах тестирования, продолжительность A/B-эксперимента должна составлять не менее 7 дней.

# Задача 2
Наша продуктовая команда в ecommerce магазине планирует запустить тест, направленный на ускорение загрузки сайта. Одна из основных метрик bounce rate в GA = 40%. Мы предполагаем, что при оптимизации сайта она изменится минимум на 20%.Средний трафик 4000 человек в день. Посчитайте сколько нам нужно дней держать эксперимент при alpha = 5% и beta = 20%

## Решение
Средний показатель отказа bounce rate составляет 40% или 0.4. Ожидаемое изменение составит 20%. Рассчитаем размер выборки при $ \alpha = 0.05, \beta = 0.20$ Для расчета времени проведения A/B теста расчитаем необходимый объем выборок для каждой группы. Для этого воспользуемся онлайн калькулятором [Evan's Awesome A/B Tools](https://www.evanmiller.org/ab-testing/sample-size.html "How many subjects are needed for an A/B test?") для сравнения конверсии:

![Расчет размера выборки](pics/pic_06_02.png "Расчет размера выборки")

Расчет показал, что объем выборки для каждой из групп должен составлять 592 пользователя, или 1184 пользователей всего. При средней посещаемости в 4000 человек в день, для проведения A/B-теста хватит одного дня.

## Ответ

1 день

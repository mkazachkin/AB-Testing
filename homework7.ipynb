# Эксперимент № 1
## Гипотеза
Размещение информации о возможности кредитования и его условиях в верхней части страницы увеличит конверсию в продажи. ICE = 360

## Что делаем
*Контрольная версия:* оставяем текущий вид страницы
*Тестовая версия:* поднимаем ссылку на страницу с информацией о возможностях кредитования и его условиях в верхнюю часть страницы.

## На каких пользователях тестируем
На всех пользователях web-версии маркетплейса. 50 % траффика.

## Сроки проведения эксперимента
Период проведения должен быть достаточно длительным, и не охватывать периоды праздников, так как в праздничные дни поведение покупателей нетипичное. Таким образом, исключаются периоды новогодних праздников, 23 февраля и 8 марта, майские праздники, 1 сентября. Сезонность также оказывает влияние. В летние месяцы на выбор объекта кредитования может сказаться период отпусков. Таким образом, наиболее подходящим периодом проведения A/B тестирования будут периоды с середины марта до конца апреля, с середины сентября до конца ноября.

Окно конверсии при покупке дорогостоящей техники, которая приобретается в т. ч. в кредит может составлять от 1 до 2 месяцев.

Длительность проведения эксперимента необходимо расчитать исходя из количества оформляеемых кредитов на сайте, а также окна конверсии. При проведении A/B-тестирования в осенний период, необходимости обрезать когорты покупателей, скорее всего, не возникнет. Длительность тестирования, проводимого в весенний период с большой вероятностью не перекроет окно конверсии всех когорт покупателей и в этом случае необходимо подобрать когорты таким образом, чтобы тестирование проводилось корректно.

Peekig Problem (Проблема подглядывания). Из-за длительного окна конверссии при покупке крупной бытовой техники и дорогой электроники на короткой дистанции метрики подвержены значительной волатильности. Необходимо полностью выдерживать срок A/B-тестирования и принимать решение уже на основе полных собранных данных.

## Предварительные работы
При необходимости, для тестирования инструментария можно провести A/A-тестирование в любой период или провести A/A тестирование на исторических данных.

## Ключевые метрики для оценки эксперимента
Количество продаж — primary. 
Количество оформленных кредитов — secondary. 

## Ожидаемый эффект
Статистически значимая разница между группами по количеству продаж в когорте покупателей средней и крупной бытовой техники и электроники среднего и верхнего ценового сегмента.

## План действий в зависимости от результатов эксперимента
При обнаружении статистически значимой разницы в пользу тестовой группы принимаем изменения в шаблон сайта. В противном случае, откатываем изменения.


# Эксперимент № 2
## Гипотеза
Выделение формы поиска с помощю анимированного или яркого элемента увеличит конверсию в использование поиска. Метрика (для метода ICE): конверсия пользователя, воспользовавшегося поиском. ICE = 350

## Что делаем
*Контрольная версия:* оставяем текущий вид страницы
*Тестовая версия:* разрабатываем новый облик формы поиска

## На каких пользователях тестируем
На всех пользователях web-версии маркетплейса. 10 % траффика.

## Сроки проведения эксперимента
Период проведения рассчитывается исходя из посещаемости сайта маркетплейса. При большой посещаемости. A/B-тестирование может быть ограничено 1 днем. Сезонность и праздничные дни на использование формы поиска влияние не оказывают. Peekig Problem (Проблема подглядывания) не возникает из-за короткого периода тестирования. В любом случае, необходимо полностью выдерживать срок A/B-тестирования и принимать решение уже на основе полных собранных данных.

## Предварительные работы
При необходимости, для тестирования инструментария и сплитования трафика можно провести A/A-тестирование в любой период или провести A/A тестирование на исторических данных.

## Ключевые метрики для оценки эксперимента
Количество пользователей воспользовавшихся формой поиска. 

## Ожидаемый эффект
Статистически значимая разница между выборками по основной метрике.

## План действий в зависимости от результатов эксперимента
При обнаружении статистически значимой разницы в пользу тестовой группы принимаем изменения в шаблон сайта. В противном случае, откатываем изменения.
